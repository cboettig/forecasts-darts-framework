```python





np.random.seed(0)
temperature = 15 + 8 * np.random.randn(2, 2, 3) # 2 x 2 x 3
lon = [-99.83, -99.32]
lat = [[42.25, 42.21], [42.63, 42.59]]
time = pd.date_range("2014-09-06", periods=3)
reference_time = pd.Timestamp("2014-09-05")

da2 = xr.DataArray(
    data=temperature,
    dims=["variable", "site", "time"],
    coords=dict(
        lon=(["site"], lon),
        lat=(["variable", "site"], lat),
        time=time,
        reference_time=reference_time,
    ),

    attrs=dict(
        description="Ambient temperature.",
        units="degC",
    ),
)
da2
```
